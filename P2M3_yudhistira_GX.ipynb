# Inisialisasi GX context

```python
context = gx.get_context()
```

# Datasource untuk CSV

```python
datasource_name = "csv-data-yudhistira"
datasource = context.sources.add_pandas(datasource_name)
```

# Data asset untuk dataset

```python
asset_name = "p2m3_yudhistira_data_clean"
asset = datasource.add_dataframe_asset(asset_name, dataframe=df)
```

# Buat batch request

```python
batch_request = asset.build_batch_request()
```

```python
print("Datasource dan Data Asset berhasil dibuat!")
print(f"Datasource name: {datasource_name}")
print(f"Asset name: {asset_name}")
```

# Great Expectations

```python
df.info()
```

```python
# Membuat Expectation Suite
suite_name = "yudhistira_data_suite"
suite = context.add_expectation_suite(expectation_suite_name=suite_name)

# Inisialisasi Validator
validator = context.get_validator(
    batch_request=batch_request,
    expectation_suite_name=suite_name,
)


# 1. Cek kolom 'id' harus unik
validator.expect_column_values_to_be_unique("id")

# 2. Cek kolom 'range_km' berada dalam rentang wajar (contoh: 50 - 1000 km)
validator.expect_column_values_to_be_between("range_km", min_value=50, max_value=1000)

# 3. Cek kolom 'battery_type' hanya berisi kategori valid
validator.expect_column_values_to_be_in_set(
    "battery_type", 
    ["Lithium-ion", "Li-ion", "NiMH", "Solid-state", "Lead-acid", "LFP", "NCM"]
)

# 4. Cek kolom 'year' bertipe numerik (karena tahun seharusnya angka)
validator.expect_column_values_to_be_in_type_list("year", ["int", "int64"])

# 5. Cek median nilai 'price_usd' berada dalam kisaran wajar
validator.expect_column_median_to_be_between("price_usd", min_value=10000, max_value=150000)

# 6. Cek kolom 'manufacturer' tidak memiliki nilai null
validator.expect_column_values_to_not_be_null("manufacturer")

# 7. Cek proporsi nilai unik di kolom 'model' dalam batas wajar
validator.expect_column_proportion_of_unique_values_to_be_between("model", min_value=0.0001, max_value=1.0)

# 8. Cek kolom 'charging_type' hanya berisi kategori valid
validator.expect_column_values_to_be_in_set(
    "charging_type", 
    ["Fast Charging", "Normal Charging", "Wireless Charging"]
)
```

**validasi yang dilakukan:** 

- to_be_unique
- to_be_between
- to_be_in_set
- to_be_in_type_list
- to_be_between (median)
- to_not_be_null
- to_be_between (unique proportion)


# Save Expectation suite

```python
# Simpan hasil expectation 
validator.save_expectation_suite(discard_failed_expectations=False)
```

# Checkpoint

```python
# Membuat atau memperbarui checkpoint
checkpoint = context.add_or_update_checkpoint(
    name="yudhistira_checkpoint",  # ubah nama agar sesuai konteks data kamu
    validations=[
        {
            "batch_request": batch_request,
            "expectation_suite_name": "yudhistira_data_suite"  # sesuaikan dengan suite yang kamu buat
        }
    ]
)

# Jalankan checkpoint untuk mengeksekusi semua expectation
checkpoint_result = checkpoint.run()

print("Checkpoint 'yudhistira_checkpoint' berhasil dijalankan!")
```

# Running

```python
result = checkpoint.run()
result.success
```
